```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data={'Fruit':['Apple','Bannana','Orange','Grape','Lemon','Pappaya'],
      'Colour':['Red','Yellow','Orange','Purple','Yellow','Orange'],}
```

```python
df=pd.DataFrame(data)
```

```python
encoder=OneHotEncoder()#it was using to change chategorical value to numarical values
```

```python
encoder
```

```python
encoder_color=encoder.fit_transform(df[['Colour']]).toarray()#after fit_transform to use to change the array form
```

```python
encoder_color
```

```python
encoder_df=pd.DataFrame(encoder_color,columns=encoder.get_feature_names_out(['Colour'])) #the get_geature_name was using the heading of the df
```

```python
encoder_df
```

```python
df_encoder=pd.concat([df,encoder_df],axis=1)#concat was the adding the dataframe
df_encoder
```
